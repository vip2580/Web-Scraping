```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
driver= webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()
driver.get("https://www.imdb.com/chart/top/")
driver.implicitly_wait(10)
```

```python
Movie_Name = []
Year_of_Release = []
Ratings =  []

names = driver.find_elements_by_xpath("//td[@class='titleColumn']")

release = driver.find_elements_by_xpath("//span[@class='secondaryInfo']")

ratings =  driver.find_elements_by_xpath("//td[@class='ratingColumn imdbRating']")


for n in names:
    Movie_Name.append(n.text)
    
for r in release:
    Year_of_Release.append(r.text)
    
for s in ratings:
    Ratings.append(s.text)
    
    
    
```

```python
driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
```
